```python
import numpy as np
import pandas as pd

import  matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('College.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print('Shape',df.shape)
print("Total number of labels: {}".format(df.shape[0]))
print("NUm of private colleges: {}".format(df[df.Private=='Yes'].shape[0]))
print("NUm of private colleges: {}".format(df[df.Private=='No'].shape[0]))
```

```python
X = df.iloc[:,1:-1]
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = df.iloc[:,0]
y
```

```python
private_encoder = LabelEncoder()

y=private_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X)
X = scaler.transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
from sklearn.svm import SVC
from sklearn import metrics

from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(X_train,y_train)
y_pred = svc_model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 'C':[0.1,1,10,100],'gamma':[1,0.1,.01,.001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

```python
grid_pred = grid.predict(X_test)
```

```python
print('Accuracy',metrics.accuracy_score(y_test,grid_pred))
```

```python
print(metrics.confusion_matrix(y_test,grid_pred))
```

```python
print(classification_report(y_test,grid_pred))
```

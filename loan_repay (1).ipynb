```python
import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
load_data = pd.read_csv('loan_borowwer_data.csv')
load_data
```

```python
load_data.info()
```

```python
target = load_data['not.fully.paid']
target
```

```python
load_data = load_data.drop(['not.fully.paid'],axis=1)
```

```python
load_data
```

```python
categorical = load_data['purpose']
```

```python
load_data = pd.get_dummies(load_data,columns=['purpose'])
```

```python
load_data
```

```python
load_data.head()
```

```python
target
```

```python
X = load_data
X
```

```python
y=target
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

print(X_train.shape)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_predict,y_test)
```

```python
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()

classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accu = accuracy_score(y_pred,y_test)
print(accu)
```
